# Plotting Learned 2D Norms

First, run the 2D norm experiments by running the command:
```
python experiment_2d_norm.py
```

The training script trains several architecture variants (Mahalanobis, Deep Norm, Wide Norm, and MLP (with ReLU activation)) on several norm types (Square ($L_\infty$), Diamond ($L_1$), random symmetric/asymmetric hulls). By default this generates results in `./2D_metrics` folder.

Then you can run the following cells to plot the results, similar to Figure 7-8 in the Appendix. The figures are saved to `./2D_norm_figs` folder. 

````python
from norm_utils import *
%matplotlib inline
````

## Small dataset, Sym/Asym Convex Hull

````python
input_folder = '2D_metrics'
datasets = ['sym_hull','asym_hull']
train_size = 16
training_sizes = [train_size]
arches = get_arches_to_plot(input_folder, training_sizes, datasets)
xlim=[-1.5,1.5]
ylim=[-1.5,1.5]

fig = plot_full_heatmap(input_folder, arches, datasets, train_size, xlim, ylim)
fig_folder = '2D_norm_figs'
if not os.path.exists(fig_folder):
    os.makedirs(fig_folder)
filename = '{}_hull_trainsize-{}'.format(input_folder, train_size)
fig.savefig(os.path.join(fig_folder, '{}.pdf'.format(filename)))
````

## Large dataset, Sym/Asym Convex Hull

````python
train_size = 128
training_sizes = [train_size]
arches = get_arches_to_plot(input_folder, training_sizes, datasets)

fig = plot_full_heatmap(input_folder, arches, datasets, train_size, xlim, ylim)
fig_folder = '2D_norm_figs'
if not os.path.exists(fig_folder):
    os.makedirs(fig_folder)
filename = '{}_hull_trainsize-{}'.format(input_folder, train_size)
fig.savefig(os.path.join(fig_folder, '{}.pdf'.format(filename)))
````
